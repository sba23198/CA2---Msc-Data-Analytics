```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Define the different parameters order to test


parametros_order = [(p, d, q) for p in range(6) for d in range(6) for q in range(6)]

# Initialise variables for best MSE and best parameter order


mejor_mse = float('inf')
mejor_order = None

# Loop over each set of parameters order


for order in parametros_order:
    # Create and fit ARIMA model with the current order parameters


    modelo = ARIMA(tweets_con_palabra_clave_BGT['sentimiento'], order=order)
    modelo_entrenado = modelo.fit()

    # Making predictions for different time horizons


    prediccion_1_dia = modelo_entrenado.forecast(steps=1)
    prediccion_3_dias = modelo_entrenado.forecast(steps=3)
    prediccion_7_dias = modelo_entrenado.forecast(steps=7)
    prediccion_1_mes = modelo_entrenado.forecast(steps=30)  # Asumiendo 30 días en un mes
    
    y_true_1_dia =  tweets_con_palabra_clave_BGT['sentimiento'].iloc[-1:]
    y_true_3_dias = tweets_con_palabra_clave_BGT['sentimiento'].iloc[-3:]
    y_true_7_dias = tweets_con_palabra_clave_BGT['sentimiento'].iloc[-7:]
    y_true_1_mes = tweets_con_palabra_clave_BGT['sentimiento'].iloc[-30:]

    # Calculate the MSE for each set of predictions


    mse_1_dia = mean_squared_error(y_true_1_dia, prediccion_1_dia)
    mse_3_dias = mean_squared_error(y_true_3_dias, prediccion_3_dias)
    mse_7_dias = mean_squared_error(y_true_7_dias, prediccion_7_dias)
    mse_1_mes = mean_squared_error(y_true_1_mes, prediccion_1_mes)

    # Calculate the total MSE by summing the MSE of all predictions.


    mse_total = mse_1_dia + mse_3_dias + mse_7_dias + mse_1_mes

    # Update the best MSE and best order parameter if applicable


    if mse_total < mejor_mse:
        mejor_mse = mse_total
        mejor_order = order

# Print the best order parameter found


print("Mejor parámetro order encontrado:", mejor_order)
print("Mejor MSE total:", mejor_mse)
```
